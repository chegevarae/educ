```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random, math
import decimal
```

```python
# 1. Задание. Нарисуйте график функции
x = np.linspace(0, 10, 130)
k = 2
a = 5
b = 9
plt.plot(x, k*np.cos((x-a)+b))
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# 3.1. Задание. Преобразование полярных координат в декартовые
rs = np.arange(0,100)
phis = np.arange(0,math.pi,math.pi/360)
xs = np.outer(rs, np.cos(phis))
ys = np.outer(rs, np.sin(phis))
print(xs, ys)
```

```python
# 3.2. Задание. Попытка нарисовать график окружности в полярных координатах :)
t = np.arange(0, 2*np.pi, 0.01)
r = 4

plt.polar(r*np.sin(t), r*np.cos(t), lw=2)                        
plt.grid(color='lightgray',linestyle='--')
plt.show()
```

```python
# 3.3. Задание. Прямая линии в полярных координатах
plt.polar([-10, 10],[0,0], color ='red', linestyle ="-")
plt.show()
```
