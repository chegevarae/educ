Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки составлял 30% от всех данных, при этом аргумент random_state должен быть равен 42.
Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model. Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.

````python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
````

````python
boston = load_boston()
boston.keys()
````

````python
data = boston["data"]
data.shape
````

````python
feature_names = boston["feature_names"]
feature_names
````

````python
# Создаем датафрейм X
X = pd.DataFrame(data, columns=feature_names)
X.head()
````

````python
# Выводим инфо по таблице
X.info()
````

````python
target = boston["target"]
target[:10]
````

````python
# Создаем датафрейм y
y = pd.DataFrame(target, columns=["price"])
y.info()
````

````python
# Разбиваем датафреймы на тренировочные и тестовые
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
````

````python
# Создаем модель линейной регрессии
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
````

````python
# Обучаем нашу модель
lr.fit(X_train, y_train)
````

````python
# Получаем предсказанные значения
y_pred = lr.predict(X_test)
y_pred.shape
````

````python
# Для проверки сопостовляем реальные значения с предсказанными
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})
check_test.head(10)
````

````python
# Оцениваем качество модели
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
````

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
````

````python
df = pd.read_csv("creditcard.csv", sep=",")
df.describe()
````

````python
df.Class.value_counts(normalize=True)
````

````python
df.info()
````

````python
pd.options.display.max_columns = 100
````

````python
df.head(10)
````

````python
# Создаем датафрейм X без столбца Class
target = "Class"
y = df[target]
X = df.drop(target, axis=1)
````

````python
X.head(10)
````

````python
# Создаем объект Series 'y' из столбца Class
y = pd.Series(y)
y.head(10)
````

````python
# чтобы 'y' стал датафреймом нужны двойные скобки [[]]
y = df[[target]]
````

````python
# Разбиваем X и y на тренировочный и тестовый наборы данных
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
````

````python
X_train.info()
````

````python
X_test.info()
````

````python
y_train.info()
````

````python
y_test.info()
````

````python
# Для поиска по сетке задаем параметры 
parameters = [{'n_estimators': [10, 15],
    'max_features': np.arange(3, 5),
    'max_depth': np.arange(4, 7)}]
````

````python
# Создаем модель GridSearchCV
clf = GridSearchCV(estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    refit=True,
    cv=3)
````

````python
# Обучаем нашу модель (несколько минут)
clf.fit(X_train, y_train)
````

````python
# Можно посмотреть документацию для best_params_
? GridSearchCV
````

````python
# Определяем лучшие параметры модели
clf.best_params_
````

````python
# Предсказываем вероятности классов
y_pred_proba = clf.predict_proba(X_test)
y_pred_proba[:10]
````

````python
# Выбераем столбец с индексом 1 (вероятность класса 1) 
y_pred_proba = y_pred_proba[:,1:2]
y_pred_proba
````

````python
# Вычисляем AUC на тестовых данных и сравниваем с результатом
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred_proba)
````
