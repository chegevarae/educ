# ДЗ 2. Анализ данных и проверка статистических гипотез

## Зависимые и независимые выборки

В чём различие между зависимыми и независимыми выборками?  

---

**Выборки называются независимыми (несвязными)**, если процедура эксперимента и полученные результаты измерения некоторого свойства у испытуемых одной выборки не оказывают влияния на особенности протекания этого же эксперимента и на результаты измерения этого же свойства у испытуемых (респондентов) другой выборки.

Примером двух независимых выборок могут служить выборки мужчин и женщин, работающих на одном предприятии (в одной отрасли и т.д.).

**Выборки называется зависимыми (связными)**, если процедура эксперимента и полученные результаты измерения некоторого свойства, проведенные на одной выборке, оказывают влияние на другую. Следует подчеркнуть, что одна и та же группа испытуемых, на которой дважды проводилось психологическое обследование (пусть даже разных психологических качеств, признаков, особенностей), по определению оказывается зависимой, или связной, выборкой.

Наиболее типичный пример зависимых выборок — повторное измерение свойства (свойств) на одной и той же выборке после воздействия (ситуация «до-после»). В этом случае выборки (одна — до экспериментального воздействия, другая — после экспериментального воздействия) зависимы в максимально возможной степени, так как они включают одних и тех же испытуемых.

## Параметрические статистические критерии и непараметрические аналоги

Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?  

---

**Параметрические критерии** включают в расчет параметры вероятностного распределения признака (средние и дисперсии).  
Примеры:  
t-критерий Стьюдента  
Критерий Фишера  
Критерий отношения правдоподобия  
Критерий Романовского  

**Непараметрические критерии** не включают в расчёт параметры вероятностного распределения и основаны на оперировании частотами или рангами.  
Примеры:  
Q-критерий Розенбаума  
U-критерий Манна — Уитни  
Критерий Уилкоксона  
Критерий Пирсона  
Критерий Колмогорова — Смирнова  

Непараметрические критерии наиболее приемлемы, когда объем выборок мал. Если данных много (например, n > 100), то не имеет смысла использовать непараметрические критерии. 

Когда выборки становятся очень большими, то выборочные средние подчиняются нормальному закону, даже если исходная переменная не является нормальной или измерена с погрешностью. 

Непараметрические тесты имеют меньшую статистическую мощность (менее чувствительны), чем параметрические, и позволяют обнаружить даже слабые отклонения.
