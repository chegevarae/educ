## Тема “Введение в математических анализ”

### Как относятся друг к другу множество и последовательность? (в ответе использовать слова типа: часть, целое, общее, частное, родитель, дочерний субъект и т.д.)

```
Последовательность это выборка элементов из множества объектов объединенных по одному признаку, она может быть равна по количеству элементов множеству или быть его частью, образуя таким образом подмножество. У последовательности есть свое множество. Оно так и называется - множество значений. Оно может быть конечным, может быть бесконечным, но это не сама последовательность.
```
